## üñºÔ∏è Visualization of the images.

```python
from htsimaging.lib.io import read_image
from htsimaging.viz.image import image_background
```

```python
## parameters
input_path='inputs/image_intensity.npy'
image_segmented_path='inputs/image_segmented.npy'
```

```python
image_intensity=read_image(input_path)
image_segmented=read_image(image_segmented_path)
```

```python
ax=image_background(
    img_region=image_segmented,
    img=image_intensity,
    cmap='gfp',
    alpha=1,
    linewidths=2,
    colors='y',
    kws_region={},
    show_scalebar=None,
    scalebar_color='w',
    test=False,
)
```
